```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import random 
import time
```

```python
url = "https://www.zillow.com/homes/for_sale/Boston,-MA_rb/"
```

```python
time_pause = 5

req_headers = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'en-US,en;q=0.8',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'
}
```

```python
time.sleep(time_pause)

with requests.Session() as ses:
    response = ses.get(url, headers = req_headers)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.title.text
```

```python
price = soup.find('div', {'class': 'StyledPropertyCardDataArea-c11n-8-82-3__sc-yipmu-0 gMDnGj'}).text
print(price)
```

```python
#dataframe for Boston, MA
web_list = list()
web_dict = {}
    
props = soup.find_all('div', {'class': 'StyledPropertyCardDataWrapper-c11n-8-82-3__sc-1omp4c3-0 iKSlva property-card-data'})
    
for x in range(0,len(props)):
    try:
        web_dict['price'] = props[x].find('div', {'class': 'StyledPropertyCardDataArea-c11n-8-82-3__sc-yipmu-0 gMDnGj'}).text
    except:
        web_dict['price'] = None
    try:
        web_dict['address'] = props[x].find('a', {'class': 'StyledPropertyCardDataArea-c11n-8-82-3__sc-yipmu-0 hiBOYq property-card-link'}).text
    except:
        web_dict['address'] = None
    try:
        web_dict['overall_size'] = props[x].find('div', {'class': 'StyledPropertyCardDataArea-c11n-8-82-3__sc-yipmu-0 eTKvfN'}).text
    except:
        web_dict['overall_size'] = None
    try:
        web_dict['brokerage'] = props[x].find('div', {'class': 'StyledPropertyCardDataArea-c11n-8-82-3__sc-yipmu-0 bJDrDR'}).text
    except:
        web_dict['brokerage'] = None
        
    web_list.append(web_dict)
    web_dict = {}
    boston_df = pd.DataFrame(web_list)
```

```python
boston_df
```

```python
boston_df['street'] = boston_df['address'].str.split(',').str[0]
boston_df['city'] = boston_df['address'].str.split(',').str[1]
boston_df['state'] = boston_df['address'].str.extract(r'([A-Z]+)(?=[^,]*$)')
boston_df['zip_code'] = boston_df['address'].str.extract(r'(\d+)(?=[^,]*$)')
#boston_df['beds'] = boston_df['overall_size'].str.extract(r'(^\d)')

boston_df
```

```python
boston_df.to_csv(r'D:\Zillow_Web_Scraping\boston_df.csv')
```
